```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"C:\Users\Lenovo\Downloads\sales_data_large.csv")
```

```python
data
```

```python
print("Total Revenue ",data["Revenue"].sum())
margin=0.3
print("Total Profit",(data["Revenue"]*margin).sum())
print("Revenue by Accessories ",data[data["Category"]=="Accessories"]["Revenue"].sum())
print("Revenue by Apparel",data[data["Category"]=="sApparel"]["Revenue"].sum())
print("Revenue by Electronics",data[data["Category"]=="Electronics"]["Revenue"].sum())
print("Revenue in first Quater of the year",data[data["Date"]<"2023-03-30"]["Revenue"].sum())
print("Revenue in Second Quater of the year",data[(data["Date"]>"2023-03-30") & (data["Date"]<"2023-06-30")]["Revenue"].sum())
print("Revenue in Third Quater of the year",data[(data["Date"]>"2023-06-30") & (data["Date"]<"2023-09-30")]["Revenue"].sum())
print("Revenue in Second Quater of the year",data[(data["Date"]>"2023-09-30") & (data["Date"]<"2023-12-31")]["Revenue"].sum())
```

```python
print("Product with Highest Price is :",data.loc[data["Price"].idxmax()]["Product"])
```

```python
category=data.groupby("Category")["Revenue"].sum()
category.plot(kind="pie",autopct="%1.1f%%",shadow=True)
plt.title("Revenue by Category")
plt.ylabel("")
plt.show()
```

```python
product=data.groupby("Product")["Revenue"].sum()
product.plot(kind="pie",autopct="%1.1f%%")
plt.ylabel("")
plt.title("Revenue by Product")
plt.show()
```

```python
data["Date"]=pd.to_datetime(data["Date"])
data["Month"]=data["Date"].dt.month_name()
month=data.groupby("Month")["Revenue"].sum()
month.plot(kind="bar")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
products=data.groupby("Product")["Quantity"].sum()
products.plot(kind="bar")
plt.xlabel("Product")
plt.ylabel("Quantity")
plt.show()
```

```python
categor=data.groupby("Category")["Quantity"].sum()
categor.plot(kind="pie",autopct="%1.1f%%")
plt.ylabel("")
plt.show()
```

```python
plt.scatter(data["Quantity"],data["Revenue"])
plt.ylabel("Revenue")
plt.xlabel("Quantity")
plt.show()
```
